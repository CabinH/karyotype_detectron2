```python
import os
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
root = '/data1/huangkaibin/Data/karyotype/origin_split/'

img_name_a = sorted([os.path.join(root, 'a', name) for name in os.listdir(os.path.join(root, 'a'))])
img_name_k = sorted([os.path.join(root, 'k', name) for name in os.listdir(os.path.join(root, 'k'))])
```

```python
ia = cv2.imread(img_name_a[0], 1)
ik = cv2.imread(img_name_k[0], 1)

fig, axes = plt.subplots(1, 2, figsize=(10, 10)) # figsize
ax1, ax2 = axes[0], axes[1] #  
# ax1.set_title("image_a")
ax1.axis("off")
ax1.imshow(ia)
# ax2.set_title("image_k")
ax2.axis("off")
ax2.imshow(ik)
plt.show()
```

```python
pos = np.where(ia!=255)
x0, x1, y0, y1 = pos[0][0], pos[0][-1], pos[1][0], pos[1][-1]
# cv2.imwrite('/data1/huangkaibin/Data/karyotype/a.png', ia[x0+1:x1, y0+1:y1])
plt.imshow(ia[x0+1:x1, y0+1:y1])
plt.show()
```
